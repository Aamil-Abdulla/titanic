```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt

from sklearn.datasets import make_circles
X,y = make_circles(n_samples=100, factor=0.3 , noise=0.1)

```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
#Dbscan
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=0.15)
predicted = dbscan.fit_predict(X)
dbscan.labels_

plt.scatter(X[:,0],X[:,1],c=predicted)
```
