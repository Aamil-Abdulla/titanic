```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_diabetes
daf = load_diabetes()
```

```python
df = pd.DataFrame(daf.data,columns=daf.feature_names)
```

```python
daf.feature_names
```

```python
plt.scatter(df.iloc[:,0],df.iloc[:,1])
```

```python
from sklearn.ensemble import IsolationForest
ifd = IsolationForest(contamination="auto")
predictions = ifd.fit_predict(df)
```

```python
import numpy as np
index = np.where(predictions<0)
index
```

```python
X = df.values
index = np.where(predictions<0)
plt.scatter(df.iloc[:,4],df.iloc[:,5])
plt.scatter(X[index,4],X[index,5],edgecolors="r")
```
