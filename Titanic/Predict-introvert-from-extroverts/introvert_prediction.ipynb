```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:/Users/Admin/Desktop/kaggle comp/Titanic/Predict-introvert-from-extroverts/train.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
x=df.drop(["id"],axis=1)
x.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df["Personality"] = le.fit_transform(df["Personality"])
df.head()
```

```python
df.shape[0]
```

```python
missing_percent = (df.isnull().sum() / df.shape[0]) * 100
print(missing_percent.sort_values(ascending=False))
```

```python
df["Drained_after_socializing"]= df["Drained_after_socializing"].fillna(df["Drained_after_socializing"].mode()[0])
df.head()
```

```python
df["Drained_after_socializing"] = le.fit_transform(df["Drained_after_socializing"])
```

```python
df.head()
```

```python

df["Stage_fear"] = le.fit_transform(df["Stage_fear"])
df.head()
```

```python
missing_percent = (df.isnull().sum() / df.shape[0]) * 100
print(missing_percent.sort_values(ascending=False))
```

```python

df["Going_outside"]=df["Going_outside"].fillna(df["Going_outside"].mean())
```

```python
missing_percent = (df.isnull().sum() / df.shape[0]) * 100
print(missing_percent.sort_values(ascending=False))
```

```python

df["Post_frequency"]=df["Post_frequency"].fillna(df["Post_frequency"].mean())
```

```python

df["Time_spent_Alone"]=df["Time_spent_Alone"].fillna(df["Time_spent_Alone"].mean())
```

```python


df["Social_event_attendance"]=df["Social_event_attendance"].fillna(df["Social_event_attendance"].mean())
```

```python
      

df["Friends_circle_size"]=df["Friends_circle_size"].fillna(df["Friends_circle_size"].mean())
df.head()
```

```python
df = df.drop(["id"], axis= 1)
```

```python
X= df.iloc[:,:-1]
X.head()
y = df["Personality"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
#RandomForest
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
model = RandomForestClassifier()
parameters = {
        'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20]
}
cv = GridSearchCV(model ,param_grid= parameters ,  scoring= "accuracy" , cv =5  )
cv.fit(X_train,y_train)
```

```python
cv.best_score_
```

```python
cv.best_params_
```

```python
#Prediction

y_pred = cv.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score , classification_report
score = accuracy_score(y_test,y_pred)
report = classification_report(y_test,y_pred)
```
